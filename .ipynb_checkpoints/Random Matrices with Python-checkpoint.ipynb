### Random Matrices with Python

**Probability Dsitributions of Random Matrix Eigenvalues**

As the name implies, random matrices are basically matrices with randomly generated elements. From a superficial viewpoint, matrices of randomly generated numbers might not be very useful, however it turns out that they have been successfully applied to various fields, such as nuclear physics and neuroscience. In this article, we will explore the properties of the distributions of the eigenvalues of random matrices!



### The Wigner Semicircle and Surmise

According to **random matrix theory**, the distribution of the eigenvalues of an ensemble of symmetric random matrices can be approximated by the Wigner semicircle:  
f(x) = (2 / (R))(R - x).

```python
import numpy as np
import matplotlib.pyplot as plt

""" Wigner Semi Circle"""
R = 2
X = np.linspace(-R, R, 1000)
semicircle = (2/(R**2 * np.pi)) * np.sqrt(R**2 - X**2)
```

Correspondingly, the distribution of eigenvalue spacings of an ensemble of symmetric random matrices can be approximated by the **Wigner surmise**:
f(s) = (s / 2)exp(-s / 4). The Wigner surmise was originally developed to describe the spaces between points in the spectra of the nuclei of heavy atoms such as Uranium.

```python
# Wigner Surmise 
s = np.linspace(0, 3, 1000)
surmise = (np.pi * s / 2) * np.exp(-np.pi * s**2 / 4)
```

### The Grand Orthogonal Ensemble

The Grand Orthogonal Ensemble uses random matrices to model Hamiltonians with time reversal symmetry.

```python
def goe_eigs( N = 10):
    # Grand Orthoganal Ensemble:
    # Hamiltonians with time reversal symmetry
    # N: size of the matrix.
    A = np.random.randn(N, N)
    B = 2**-0.5 * (A + A.T)
    D = np.linalg.eigvals(B)
    return np.real(D)
```

### The Grand Unified Ensemble

The Grand Unified Ensemble uses Hamiltonians without time reversal symmetry.

```python
def gue_eigs(N=10):
    # Grand Unified Ensemble:
    # Hamiltonians without time reversal symmetry
    # N : size of the matrix.
    A = np.random.randn(N,N) + 1j * np.random.randn(N,N)
    B = 2**-1 * (A + np.conj(A.T))
    D = np.linalg.eigvals(B)
    return np.real(D)
```

### The Symmetric Bernoulli Ensemble

Finally, we have the Symmetric Bernoulli Ensemble which are random sign matrices constrained to be symmetric.

```python
def bernoulli_eigs( N=10):
    # Symmetric Bernoulli Ensemble
    # Random Sign matrices which are constrained to be symmetric.
    # N: size of the matrix.
    B = np.zeros([N,N])
    for i in range(N):
        for j in range(N):
            if i >= j & i != j:
                x = np.random.random()
                if x<= 0.5:
                    B[i][j] = 1
                    B[j][i] = 1
                else:
                    B[i][j] = -1
                    B[j][i] = -1
    D = np.linalg.eigvals(B)
    return np.real(D)
```

### Eigenvalues of the Grand Orthogonal Ensemble

For now we will explore the eigenvalue distribution of the Grand Orthogonal Ensemble. Using the code below, we generate 1000 different symmetric random matrices with dimensions of 10 X 10, and compute and plot the spacings between the eigenvalues. 

```python
def random_matrix_eigenvals(randmatfunc = goe_eigs,
                            n_matrices = 1000,
                            N = 10):
    # randmatfunc: random matrix function.
    # n_matrices: number of random matrices to generate
    # N: matrix size
    
    # Generate eigenvalues using randmatfunc:
    eig = np.zeros([n_matrices, N])
    for i in range(n_matrices):
        eig[i,:] = randmatfunc(N)
    # Normalizing the eigenvalues
    eig = eig / np.sqrt(N)
    
    # Obtain the eigenvalue spacing:
    # diff and sort operates row-wise by default
    spacing = np.diff(np.sort(eig))
    # np.mean operates over all matrix elements
    spacing = spacing / np.mean(spacing)
    
    return eig, spacing

def plot_random_matrix_eigenvalues(randmatfunc= goe_eigs,
                              n_matrices = 1000, N = 10):
    eig, spacing = random_matrix_eigenvals(randmatfunc, n_matrices, N)
    
    plt.figure(figsize=(15,5))
    plt.subplot(1,2,1)
    plt.title("Random Matrix Eigenvalue")
    plt.hist(eig.ravel(), 100, density=True)
    plt.plot(X, semicircle)
    plt.grid(True)
    plt.legend(["Wigner Semicircle", "Eigenvalues"])
    plt.subplot(1,2,2)
    plt.title("Random Matrix Eigenvalue Spacing")
    plt.hist(spacing.ravel(), 100, density=True)
    plt.plot(s, surmise)
    plt.grid(True)
    plt.legend(["Wigner Surmise", "Eigenvalue Spacings"])
    plt.show()
```

For the Grand Orthogonal ensemble, we find that the distribution of the eigenvalues are indeed approximated by the Wigner semicircle, and the distribution of the eigenvalue spacings are approximated by the Wigner Surmise.

```python
plot_random_matrix_eigenvalues(randmatfunc= goe_eigs,
                               n_matrices = 1000,
                               N = 100)
```

### Eigenvalues of the Grand Unified Ensemble and the Symmetric Bernoulli Ensemble

Likewise, we find that the Wigner Semicircle and surmise are good approximations for both the Grand Unified Ensemble and the Symmetric Bernoulli Ensemble as well. 

```python
plot_random_matrix_eigenvalues(randmatfunc = gue_eigs,
                               n_matrices = 1000,
                               N = 100)
```
