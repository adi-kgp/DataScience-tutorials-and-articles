## Monte Carlo Simulation for Engineering Transfer Functions

One use of Monte Carlo Simulation is to predict the outcome of a transfer function in a tolerance analysis. Let's do the analysis.

```python
# Import all the libraries
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

### Transfer function Definition

Lets use a very simple function given by the following equation:

y = 20A + 5B

The implementation in python is as follows:

```python
# transfer function definition
def transfer_function(A,B):
    return 20*A + 5*B
```

### Monte Carlo Function Definition
The idea is to randomly select the value out of a distributions assigned to variables in transfer function and then calculate the value of the function. For this example let us assign a normal distribution to both variables A and B.

```python
# Monte Carlo Simulation function
def Monte_Carlo(transfer_function, iterations):
    final_results = []
    for n in range(iterations):
        a = random.gauss(200, 33)
        b = random.gauss(50, 5)
        final_results.append(transfer_function(a,b))
    return final_results
```

The number of iterations is controlled via variable iterations. For this example, variable **a** has been assigned normal distribution with mean value of 200 and a standard deviation of 33. Variable **b** has been assigned normal distribution with mean value of 50 and a standard deviation of of 5. 

Let's check the number of iterations influences the result of Monte Carlo simulation. 

### Running the simulation

Once all is defined the Monte Carlo simulation is one line of code. In this example simulation with 10000 iterations has been saved in the variable **k**, and the next one with just 100 iterations has been saved in the variable **l**. Please see the code below: 

```python
# number of iterations
iterations = 10000
k = Monte_Carlo(transfer_function, iterations=iterations)
l = Monte_Carlo(transfer_function, iterations=100)
```

The comparison of the results can be plotted through following code

```python
fig = plt.figure(figsize=(10,5))
sns.distplot(k, label="10000 iter")
sns.distplot(l, label = "100 iter")
plt.legend()
plt.title("Monte Carlo Simulation");
```
